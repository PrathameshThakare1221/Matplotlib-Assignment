```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])

axes.plot(x,y,'g','--')
axes.set_xlabel('X Axis Values')
axes.set_ylabel('Y Axis Values')
axes.set_title('Graph X vs Y')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,'r')
ax1.set_xlabel('X Axis Values ')
ax1.set_ylabel('Y Axis Values')
ax1.set_title('Graph X vs Y')

ax2.plot(x,y,'g')
ax2.set_xlabel('X Axis Values ')
ax2.set_ylabel('Y Axis Values')
ax2.set_title('Graph X vs Y')

```

```python
fig=plt.figure()
ax3=fig.add_axes([0,0,1,1])
ax4=fig.add_axes([0.2,0.5,.4,.4])

ax3.plot(x,y,'b')
ax3.set_xlabel('X Axis Values ')
ax3.set_ylabel('Y Axis Values')
ax3.set_title('Graph X vs Y')

ax4.plot(x,y,'g')
ax4.set_xlabel('X Axis Values ')
ax4.set_ylabel('Y Axis Values')
ax4.set_title('Graph X vs Y')

```

```python
plt.subplots(nrows=1, ncols=2)
```

```python
fig,ax = plt.subplots(figsize=(10,5))

ax.plot(x,x+1,color='green',linewidth=0.25)
ax.plot(x,y,'--')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('X and Y')
```

```python
fig,ax = plt.subplots(figsize=(12,6))

ax.plot(x,color='m',linewidth=0.50)
ax.plot(x,y,'--')
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_title('X and Z')
```
